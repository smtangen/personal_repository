```python
import pandas as pd
import numpy as np

def load_data():
    data = pd.read_csv("FantasyPros_2018_Week_1_FLEX_Rankings.csv")
    data["Std Dev"].replace(0.0, None, inplace=True)
    data["Score"] = data["Avg"] * data["Avg"] * np.sqrt(data["Std Dev"])
    data.sort_values(by="Score", inplace=True)
    data["Pos_Name"] = data.Pos.str[:2]
    
    return data[(data["Best"] <= 150) & (data["Team"] != "FA")]


def load_MSBA_team():
    MSBA_team = pd.read_csv("MSBA_team.csv")
    
    return MSBA_team

def load_family_team():
    family_team = pd.read_csv("family_team.csv")
    
    return family_team


def flex():
    data = load_data() 
    MSBA_team = load_MSBA_team()
    family_team = load_family_team()
    data_MSBA = data[data["Flex Position (RB/WR/TE)"].isin(MSBA_team["Player"])]
    data_family = data[data["Flex Position (RB/WR/TE)"].isin(family_team["Player"])]
    
    return data_MSBA  # switch based on which lineup I want to look at


def run_b():
    data = load_data()
    
    return data[data["Pos_Name"] == "RB"].head(30)


def wide_r():
    data = load_data()
    
    return data[data["Pos_Name"] == "WR"].head(30)


def tight_end():
    data = load_data()
    
    return data[data["Pos_Name"] == "TE"].head(25)


def qb_data():
    qb_data = pd.read_csv("FantasyPros_2018_Week_1_QB_Rankings.csv")
    qb_data["Std Dev"].replace(0.0, None, inplace=True)
    qb_data["Score"] = qb_data["Avg"] * qb_data["Avg"] * np.sqrt(qb_data["Std Dev"])
    qb_data.sort_values(by="Score", inplace=True)
    
    return qb_data[qb_data["Team"] != "FA"].head(32)


def dst_data():
    dst_data = pd.read_csv("FantasyPros_2018_Week_1_DST_Rankings.csv")
    dst_data["Std Dev"].replace(0.0, None, inplace=True)
    dst_data["Score"] = dst_data["Avg"] * dst_data["Avg"] * np.sqrt(dst_data["Std Dev"])
    dst_data.sort_values(by="Score", inplace=True)
    
    return dst_data
    

def k_data():
    k_data = pd.read_csv("FantasyPros_2018_Week_1_K_Rankings.csv")
    k_data["Std Dev"].replace(0.0, None, inplace=True)
    k_data["Score"] = k_data["Avg"] * k_data["Avg"] * np.sqrt(k_data["Std Dev"])
    k_data.sort_values(by="Score", inplace=True)
    
    return k_data[k_data["Team"] != "FA"].head(32)
```
